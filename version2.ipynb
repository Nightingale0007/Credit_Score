> ## Data preprocessing

```python
import pandas as pd
import numpy as np
#Importing data and marking null values
def load_data():
    return pd.read_csv(filepath_or_buffer = "./train.csv",
                       na_values = ['','_______','_','NA','!@9#%8','#F%$D@*&8','NM','__10000__'],
                       low_memory = False,
                       nrows= 100000
                       )
origen_df = load_data()
copy_df = origen_df.copy()
string_columns = ['Name','SSN','Occupation','Type_of_Loan','Credit_Mix','Credit_History_Age','Payment_of_Min_Amount','Payment_Behaviour']
number_columns = ['Age','Annual_Income','Monthly_Inhand_Salary','Num_Bank_Accounts','Num_Credit_Card',
                            'Interest_Rate','Num_of_Loan','Delay_from_due_date','Num_of_Delayed_Payment','Changed_Credit_Limit',
                            'Num_Credit_Inquiries','Outstanding_Debt','Credit_Utilization_Ratio','Total_EMI_per_month','Amount_invested_monthly',
                            'Monthly_Balance']
#Mark strings in numeric types as null values
def mark_strings_as_nan(df, columns):
    for column in columns:
        df[column] = pd.to_numeric(df[column], errors='coerce')
#Mark negative numbers as null
def mark_negatives_as_nan(df, columns):
        for column in columns:
            if(column != 'Delay_from_due_date' and column != 'Changed_Credit_Limit'):
                df[column] = np.where(df[column] < 0, np.nan, df[column])
#Marks numbers outside the valid range as nulls
def mark_outliers_as_nan(df, columns):
    for column in columns:
        if(column != 'Amount_invested_monthly' and column != 'Delay_from_due_date' and column != 'Changed_Credit_Limit' and column != 'Credit_Utilization_Ratio'
           and column != 'Monthly_Balance'):
            median = df[column].median()
            mad = np.median(np.abs(df[column] - median))
            
            lower_limit = median - (3 * mad)
            upper_limit = median + (3 * mad)
            
            df[column] = np.where((df[column] < lower_limit) | (df[column] > upper_limit), np.nan, df[column])
    df['Age'] = np.where(df['Age'] > 94, np.nan, df['Age'])
# Set the null value of a string type to the value with the highest number of occurrences, averaged for a specific column
def fill_most_frequent(df, columns):
    for column in columns:
        # if column == 'Num_of_Delayed_Payment' or column == 'Changed_Credit_Limit' or column == 'Amount_invested_monthly':
        #     # For 'Num_of_Delayed_Payment' and 'Changed_Credit_Limit', take the average value
        #     df[column].fillna(df[column].mean(), inplace=True)
        # else:
            # For the other columns, take the value with the highest number of occurrences
        mode_values = df[column].mode()
        if not mode_values.empty:
            mode_value = mode_values.iloc[0]
            df[column].fillna(mode_value, inplace=True)
        
df_p = copy_df[copy_df['Credit_Score'] == 'Poor'].copy()
df_s = copy_df[copy_df['Credit_Score'] == 'Standard'].copy()
df_g = copy_df[copy_df['Credit_Score'] == 'Good'].copy()
list_of_dfs = [df_p,df_s,df_g]

for df in list_of_dfs:
    mark_strings_as_nan(df,number_columns)
    mark_negatives_as_nan(df,number_columns)
    mark_outliers_as_nan(df,number_columns)
    fill_most_frequent(df,number_columns + string_columns)
new_df = pd.concat(list_of_dfs, ignore_index=True)
```

> # Decision Tree

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.classification import DecisionTreeClassifier

spark = SparkSession.builder.appName("RandomForest").getOrCreate()
data = spark.read.csv("G5output.csv", header=True, inferSchema=True)
data = data.withColumn("label", data["Credit_Score"])
feature_cols = data.columns[:-2]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
assembled_data = assembler.transform(data)

train_data, test_data = assembled_data.randomSplit([0.8, 0.2], seed=1)

# Creating a Decision Tree Classifier
classifier = DecisionTreeClassifier(labelCol="label", featuresCol="features", maxDepth=10)

model = classifier.fit(train_data)
predictions = model.transform(test_data)
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Decision Tree Model Accuracy (before fine-tune): {accuracy}")


param_grid = ParamGridBuilder() \
    .addGrid(classifier.maxDepth, [5, 10, 15,20]) \
    .addGrid(classifier.maxBins, [8,16,32,64]) \
    .build()

# Creating a cross-validator
cross_validator = CrossValidator(estimator=classifier,
                                 estimatorParamMaps=param_grid,
                                 evaluator=MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy"),
                                 numFolds=3)
# Perform model fitting
cv_model = cross_validator.fit(train_data)
best_model = cv_model.bestModel
# Predictions on test data
predictions = best_model.transform(test_data)

# Evaluating Performance with the Evaluator
accuracy = evaluator.evaluate(predictions)
print(f"Decision Tree Model Accuracy (After fine-tune): {accuracy}")

# Obtaining hyperparameters for the best model
best_max_depth = best_model.getMaxDepth()
best_max_bins = best_model.getMaxBins()

print(f"Best Max Depth: {best_max_depth}")
print(f"Best Max Bins: {best_max_bins}")
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

spark = SparkSession.builder.appName("Bayes").getOrCreate()
data = spark.read.csv("G5output.csv", header=True, inferSchema=True)
#Selection of data based on relevance
choice = data.select('Num_Credit_Card','Interest_Rate','Delay_from_due_date','Credit_Mix','Credit_History_Age','Payment_of_Min_Amount',
                     'Credit_Score')
#Renaming columns
choice = choice.withColumnRenamed("Credit_Score","label")

# Creating feature vectors
feature_columns = choice.columns[:-1] 
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
data_with_features = assembler.transform(choice)

# Divide the training set and test set
train_data, test_data = data_with_features.randomSplit([0.8, 0.2], seed=42)

# pre-moderation
nb = NaiveBayes(labelCol="label", featuresCol="features", smoothing=0.1)

# Training Models
model = nb.fit(train_data)

# Predictions on the test set
predictions = model.transform(test_data)

# Evaluating model performance
evaluator = MulticlassClassificationEvaluator(metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Accuracy:", accuracy)

# After the call 
# Create NaiveBayes classifiers
nb = NaiveBayes(labelCol="label", featuresCol="features")

# Creating a Parametric Grid
paramGrid = ParamGridBuilder() \
    .addGrid(nb.smoothing, [0.1, 0.5, 1.0]) \
    .build()

# Creating a cross-validation evaluator
evaluator = MulticlassClassificationEvaluator(metricName="accuracy")
crossval = CrossValidator(estimator=nb,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=5)

# Training Models
model = crossval.fit(train_data)

# Predictions on the test set
predictions = model.bestModel.transform(test_data)

# Evaluating model performance
accuracy = evaluator.evaluate(predictions)
print("Accuracy:", accuracy)

# Getting the parameters of the best model
bestModel = model.bestModel
bestSmoothing = bestModel.getOrDefault('smoothing')
print("Best Smoothing:", bestSmoothing)
```

> ## Read csv file

```python
# create spark session
spark = SparkSession.builder.appName("Random-Forest").getOrCreate()

# read the csv file
data = spark.read.csv("G5output.csv", header = True, inferSchema = True)
```

```python
# print the schema of the data frame
data.printSchema()
```

the data type of the columns are all numerical type

```python
# check current data
pd.DataFrame(data.take(5), columns = data.columns).transpose()
```

> ## Pre-processing

```python
# create new column "features" that combines all the value of feature columns into a vector
feature_columns = data.columns[:-2]
assembler = VectorAssembler(inputCols = feature_columns, outputCol = "features")
data = assembler.transform(data)

# change the column name of the target column
data = data.withColumnRenamed("Credit_Score", "label")
```

```python
# check the data
pd.DataFrame(data.take(5), columns = data.columns).transpose()
```

> ## Stratified Sampling

```python
# get unique labels of the target column
labels = data.select("label").distinct().collect()
# define parameter "fractions" that will be used in "SampleBy"
fractions = {label.__getitem__("label") : 0.8 for label in labels} # fractions will be: {1:0.8, 2:0.8, 3:0.8}

# split the data into 80% training set and 20% test set
train_set = data.sampleBy(data["label"], fractions, seed=0)
test_set = data.subtract(train_set)
```

```python
# check the result of training set
print("Total number of records in training set: ", train_set.count())
print("Class distributions: ", train_set.groupBy("label").count().collect())
```

```python
# check the result of test set
print("Total number of records in test set: ", test_set.count())
print("Class distributions: ", test_set.groupBy("label").count().collect())
```

The number of records and the class distributions of training set and test set approximately conform with the required proportion.

> ## Standard Random Forest Model

>> ### Training

```python
# initialize Random Forest Classifier
rfc = RandomForestClassifier(featuresCol = "features", labelCol = "label")
rf_model = rfc.fit(train_set)
```

>> ### Evaluation

```python
# test on the model
predictions_standard = rf_model.transform(test_set)

# evaluate the model
evaluator = MulticlassClassificationEvaluator(labelCol = "label", predictionCol = "prediction", metricName="accuracy")
accuracy_standard = evaluator.evaluate(predictions_standard)

# show the result
print("Standard Model: ")
print("Accuracy: ", accuracy_standard)
```

> ## Fine-tuning

>> ### Training

```python
# initialize hyper-parameter grid
para_grid = ParamGridBuilder() \
    .addGrid(rfc.numTrees, [10, 20, 30]) \
    .addGrid(rfc.maxDepth, [10, 15, 20]) \
    .build()

# initialize cross validator
cross_validator = CrossValidator(estimator = rfc, 
                                estimatorParamMaps = para_grid, 
                                evaluator = MulticlassClassificationEvaluator(labelCol="label", 
                                                                              predictionCol = "prediction", 
                                                                              metricName="accuracy"), 
                                numFolds = 5, 
                                seed = 0)

# Train the model
cv_model = cross_validator.fit(train_set)
```

>> ### Evaluation

```python
# get the best model
best_model = cv_model.bestModel

# predict on test set
predictions_best = best_model.transform(test_set)

# evaluate the model
accuracy_best = evaluator.evaluate(predictions_best)

# show the result
print("Best Model: ")
print("Accuracy: {:.2f}".format(accuracy_best))
print("Best number of trees: {}".format(best_model.getNumTrees))
print("Best max depth: {}".format(best_model.getMaxDepth()))
```
